## Equilibrium for each graph

```python
import time
num_t = 5
ll_t = [np.linspace(2,4.5,num=num_t),np.linspace(2,4,num=num_t),np.linspace(2,4,num=num_t),np.linspace(12,25,num=num_t)]
L_max_flip = [10000]*(len(L_Graph)-1) + [20000]
n_seq = 1000

l_graph_temp =[]
l_graph_flip = []
M_msa_hist = np.zeros((len(L_Graph),num_t,n_seq))
for i in range(len(L_Graph)):
    Graph = L_Graph[i]
    MSA = ge.Sampling_msa_two_states(Graph)
    max_flip = L_max_flip[i]
    delta_flip = int(max_flip/100)
    l_t_plot = []
    tot_time = 0
    l_t = ll_t[i]
    for i_t,t in enumerate(l_t):
        MSA.T = t
        msa = np.random.randint(2,size=(n_seq,Graph.number_of_nodes()),dtype = np.int8)*2-1
        l_plot = [np.mean(np.abs(np.mean(msa,axis=1)))]
        l_flip = [0]
        c = 0
        while c<max_flip:
            for j in range(msa.shape[0]):
                MSA.mcmc(delta_flip,msa[j])
            c+=delta_flip
            l_plot.append(np.mean(np.abs(np.mean(msa,axis=1))))
            l_flip.append(c)
        M_msa_hist[i,i_t] = np.mean(msa,axis=1)
        l_t_plot.append(l_plot)
    l_graph_flip.append(l_flip)
    l_graph_temp.append(l_t_plot)
```

```python
figsize=set_size(width,subplots=(2,2),scale_height=1)
plt.figure(figsize=figsize)
fig, ax =  plt.subplots(figsize=figsize, ncols=2, nrows=2)
for i_graph,graph in enumerate(L_Graph):
    l_t = ll_t[i_graph]
    for i,t in enumerate(l_t):
        l_t_plot =  l_graph_temp[i_graph]
        ax[i_graph//2,i_graph%2].plot(l_graph_flip[i_graph],l_t_plot[i],label="T=%s"%np.round(t,1))
        if i_graph==0:
            ax[i_graph//2,i_graph%2].set_xlabel("Mutations accepted")
            ax[i_graph//2,i_graph%2].set_ylabel("Abs Magnetisation")
        ax[i_graph//2,i_graph%2].set_title(str(L_Graph[i_graph]))
        #plt.semilogx()
fig.tight_layout()
for i_graph,graph in enumerate(L_Graph):
    ax[i_graph//2,i_graph%2].legend(bbox_to_anchor=(0.55, -0.05), loc=3 )
fig.savefig("figures_different_graphs/Magentisation_graph.svg")
```

```python
figsize=set_size(width,subplots=(len(L_Graph),len(l_t)),fraction=2)
plt.figure(figsize=figsize)
for i in range(len(L_Graph)):
    Graph = L_Graph[i]
    l_t = ll_t[i]
    for i_t,t in enumerate(l_t):
        plt.subplot(len(L_Graph),len(l_t),(i_t+1) + (i)*len(l_t))
        if i_t == 0:
            plt.ylabel(str(Graph),fontsize=18)
        plt.title("T = %s"%(np.round(t*100)/100),fontsize=18)
        bins = np.arange(-1,1,16/Graph.number_of_nodes())
        plt.hist(M_msa_hist[i,i_t],bins=bins,range=(-1,1))
plt.tight_layout()
plt.savefig("figures_different_graphs/histo_magmetisation.svg")
```

# Global parameter

```python
l_tc = [3.8 , 2.7 , 2.8 , 20] ## List of the critical temperature
## Temperature ####
LL_T = [np.linspace(0.5*tc,2.5*tc, num = 20) for tc in l_tc]

## Mutation per generation ####
prob_mutation_site = 30/200
n_generations = 10 # 2^10 = 1024 chain of spin
n_avg = 10
n_seq = int(2**n_generations)
do_wolf_flip = 0
```

# Generation of Sequence

```python
save = True
if save:
    l_graph_temp_no_phylo = []
    l_graph_temp_phylo_rd = []
    l_graph_temp_phylo_equi = []
    l_graph_msa_pure_phylo = []
    
    for ind_Graph,Graph in enumerate(L_Graph):
        print("Graph : " + str(ind_Graph) )
        sampling = ge.Sampling_msa_two_states(Graph, flip_equi=L_max_flip[ind_Graph]) 

        LT = LL_T[ind_Graph]
        n_mut_gen = int(prob_mutation_site*L_Graph[ind_Graph].number_of_nodes())
        print("%s : %s mutations per generation"%(str(ind_Graph),n_mut_gen))

        l_graph_temp_no_phylo.append( loop.loop_sampling_temperature_wolf(n_avg,LT,sampling,sampling.msa_no_phylo,n_seq,do_wolf_flip) )
        l_graph_temp_phylo_rd.append( loop.loop_sampling_temperature_wolf(n_avg,LT,sampling,sampling.msa_binary_tree,n_generations,n_mut_gen,0,do_wolf_flip) )
        l_graph_temp_phylo_equi.append( loop.loop_sampling_temperature_wolf(n_avg,LT,sampling,sampling.msa_binary_tree,n_generations,n_mut_gen,1,do_wolf_flip) )

        l_graph_msa_pure_phylo.append(loop.loop_on_function(n_avg,sampling.msa_binary_tree_pure_phylo,n_generations,n_mut_gen,0))

    with open("data_diff_graph/l_graph_temp_no_phylo.txt", "wb") as fp:   #Pickling
        pickle.dump(l_graph_temp_no_phylo, fp)
    with open("data_diff_graph/l_graph_temp_phylo_rd.txt", "wb") as fp:   #Pickling
        pickle.dump(l_graph_temp_phylo_rd, fp)
    with open("data_diff_graph/l_graph_temp_phylo_equi.txt", "wb") as fp:   #Pickling
        pickle.dump(l_graph_temp_phylo_equi, fp)
    with open("data_diff_graph/l_graph_msa_pure_phylo.txt", "wb") as fp:   #Pickling
        pickle.dump(l_graph_msa_pure_phylo, fp)
```

```python
load = True
if load:
    with open("data_diff_graph/l_graph_temp_no_phylo.txt", "rb") as fp:   #Pickling
        l_graph_temp_no_phylo = pickle.load(fp)
    with open("data_diff_graph/l_graph_temp_phylo_rd.txt", "rb") as fp:   #Pickling
        l_graph_temp_phylo_rd = pickle.load(fp)
    with open("data_diff_graph/l_graph_temp_phylo_equi.txt", "rb") as fp:   #Pickling
        l_graph_temp_phylo_equi = pickle.load(fp)
    with open("data_diff_graph/l_graph_msa_pure_phylo.txt", "rb") as fp:   #Pickling
        l_graph_msa_pure_phylo = pickle.load(fp)
```

## Impact of the temperature on several graph

```python
l_middle_index = [np.argwhere(np.diff([G.nodes[node]["subset"] for node in range(G.number_of_nodes())])==True)[0,0]+1 for G in L_Graph]

kwargs = {"s_train" : s_train,
         "reg": pseudo_count,
          "n_pair" : n_pair_species,
          "theta" : theta,
          "fast" : True,
          "l_middle_index" : l_middle_index
         }

print(kwargs)
l_no_phylo_temp = loop.loop_inf_partners_diff_graph(inf.Inference_Partner,l_graph_temp_no_phylo,**kwargs)
l_binary_tree_rand_temp = loop.loop_inf_partners_diff_graph(inf.Inference_Partner,l_graph_temp_phylo_rd,**kwargs)
l_binary_tree_equi_temp = loop.loop_inf_partners_diff_graph(inf.Inference_Partner,l_graph_temp_phylo_equi,**kwargs)

tree_rand_phylo = []
for l_msa_pure_phylo in l_graph_msa_pure_phylo:
    tree_rand_phylo.append(inf.Inference_Partner(l_msa_pure_phylo, s_train, pseudo_count, n_pair_species, theta=theta, fast=True, middle_index=None))
```

```python
import matplotlib.lines as mlines
label_0 = "Independent seq."
label_1 = "Phylo tree\nrand. ancestor"
label_2 = "Phylo tree\nequi. ancestor"
figsize=set_size(width,subplots=(1,2),scale_height=1.2)
fig, ax =  plt.subplots(figsize=figsize, ncols=2)
for i in range(len(LL_T)):
    max_i = np.argmax(l_no_phylo_temp[i])
    tm = LL_T[i][max_i]
    ax[i//2].plot(LL_T[i]/tm,l_no_phylo_temp[i], ls = "-", color=cmaps[i], label = "Independent sequences : \n %s"%(str(L_Graph[i])))
    ax[i//2].plot(LL_T[i]/tm,l_binary_tree_rand_temp[i], ls = "dotted", color=cmaps[i], label ="Binary tree random ancestor : \n %s"%(str(L_Graph[i])))
    ax[i//2].plot(LL_T[i]/tm,l_binary_tree_equi_temp[i], ls = "--", color=cmaps[i], label = "Binary tree equilibrium ancestor : \n%s"%(str(L_Graph[i])))
ax[0].set_xlabel("$\\frac{T}{T_{max}}$")
ax[1].set_xlabel("$\\frac{T}{T_{max}}$")
ax[0].set_ylabel("Fraction of correct predictions")
#ax.semilogx()
ax[0].grid(which='both')
ax[1].grid(which='both')
ax[0].get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())

Lhandles = []
Lhandles.append( mlines.Line2D([], [], color=cmaps[0], label="Erdos" ))
Lhandles.append( mlines.Line2D([], [], color=cmaps[1], markersize=15, label="SBM"))
Lhandles.append( mlines.Line2D([], [], color=cmaps[2], markersize=15, label="Interface"))
Lhandles.append( mlines.Line2D([], [], color=cmaps[3], markersize=15, label="HK-RR"))
Lhandles.append(mlines.Line2D([],[],ls = '-', color = "black", label = label_0))
Lhandles.append(mlines.Line2D([],[],ls = '--',color = "black", label = label_1))
Lhandles.append(mlines.Line2D([],[],ls = 'dotted',color = "black",lw = 2, label = label_2))

ax[1].legend(bbox_to_anchor=(1, 1),loc='best', ncol=1, handles=Lhandles,markerfirst=True)
fig.tight_layout()
fig.savefig("figures_different_graphs/Partners_temperature_different_graph_tmax.svg",bbox_inches = "tight")
plt.show()
```
