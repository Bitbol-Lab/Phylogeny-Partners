```python
import math 
import random
def Interface_Graph(sizes, p_in, p_out, number_node_in_contact, seed=None):
    """Return the random partition graph with a partition of sizes.

    A partition graph is a graph of communities with sizes defined by
    s in sizes. Nodes in the same group are connected with probability
    p_in and nodes of different groups are connected with probability
    p_out.
    
    Parameters
    ----------
    sizes : list of ints
      Sizes of groups
    p_in : float
      probability of edges with in groups
    p_out : float
      probability of edges between groups
    number_node_in_contact : int
        number of node of each block which are going to form link with other group
    directed : boolean optional, default=False
      Whether to create a directed graph
    seed : int optional, default None
      A seed for the random number generator

    Returns
    -------
    G : NetworkX Graph or DiGraph
      random partition graph of size sum(gs)

    Raises
    ------
    NetworkXError
      If p_in or p_out is not in [0,1]

       """
    # Use geometric method for O(n+m) complexity algorithm
    # partition=nx.community_sets(nx.get_node_attributes(G,'affiliation'))
    directed=False
    if not seed is None:
        random.seed(seed)
    if not 0.0 <= p_in <= 1.0:
        raise nx.NetworkXError("p_in must be in [0,1]")
    if not 0.0 <= p_out <= 1.0:
        raise nx.NetworkXError("p_out must be in [0,1]")

    G = nx.Graph()
    G.name = "Interface"
    G.graph['partition'] = []
    n = sum(sizes)
    G.add_nodes_from(range(n))
    # start with len(sizes) groups of gnp random graphs with parameter p_in
    # graphs are unioned together with node labels starting at
    # 0, sizes[0], sizes[0]+sizes[1], ...
    next_group_false = {}  # maps node key (int) to first node in next group
    start = 0
    start_false = 0
    group = 0
    for n in sizes:
        edges = ((u+start, v+start)
                 for u, v in
                 nx.fast_gnp_random_graph(n, p_in, directed=directed).edges())
        G.add_edges_from(edges)
        next_group_false.update(dict.fromkeys(range(start_false, start_false+number_node_in_contact), start_false+number_node_in_contact))
        G.graph['partition'].append(set(range(start, start+n)))
        group += 1
        start_false += number_node_in_contact
        start += n
    L_sizes = [0]
    for i in range(len(sizes)-1):
        L_sizes.append(L_sizes[i]+sizes[i])
    ###
    # handle edge cases
    if p_out == 0:
        return G
    if p_out == 1:
        print(next_group_false)
        for n in next_group_false:
            targets = range(next_group_false[n], len(sizes)*number_node_in_contact)
            n = n//number_node_in_contact*L_sizes[n//number_node_in_contact] + n%number_node_in_contact
            New_targets = []
            for i in targets:
                New_targets.append( i//number_node_in_contact*L_sizes[i//number_node_in_contact] + i%number_node_in_contact ) 
            G.add_edges_from(zip([n]*len(New_targets), New_targets))
        return G
    # connect each node in group randomly with the nodes not in group
    # use geometric method like fast_gnp_random_graph()
    ### modif

    lp = math.log(1.0 - p_out)
    n = number_node_in_contact*len(sizes)#len(G)
    for u in range(n-1):
        v = next_group_false[u]  # start with next node not in this group
        while v < n:
            lr = math.log(1.0 - random.random())
            v += int(lr/lp)
            if v < n:
                G.add_edge(u//number_node_in_contact*L_sizes[u//number_node_in_contact] + u%number_node_in_contact, v//number_node_in_contact*L_sizes[v//number_node_in_contact] + v%number_node_in_contact)
                v += 1
    return G
```

```python
%%cython 

import numpy as np
cimport cython
from libc.stdlib cimport RAND_MAX
from libc.math cimport exp
from libc.time cimport time
from posix.stdlib cimport random, srandom, drand48

cdef class Creation_MSA_Generation:
    cdef int[:,:] M_neighboors
    cdef int[::1] l_n_neighboors
    cdef int Number_of_Node
    cdef public float Temperature
    cdef public int Flip_equilibrium_standard_algo
    cdef public int Flip_equilibrium_wolf_algo
    cdef public int wolf
    cdef int Number_Generation
    cdef int Numb_Mutation_Per_Generation
    cdef int Number_for_average
    
    def __init__(self, Graph, int Number_Generation, int Numb_Mutation_Per_Generation, int Number_for_average,
                 int Flip_equilibrium_standard_algo = 4000, int Flip_equilibrium_wolf_algo = 20 , wolf = False  ):
        """
        Graph
        Number_Generation
        Numb_Mutation_Per_Generation
        Number_for_average
        print_progress
        """
        self.Number_of_Node = np.intc(Graph.number_of_nodes())
        self.Number_for_average = Number_for_average
        self.Number_Generation = Number_Generation
        self.Numb_Mutation_Per_Generation = Numb_Mutation_Per_Generation
        self.Flip_equilibrium_standard_algo = Flip_equilibrium_standard_algo
        self.Flip_equilibrium_wolf_algo = Flip_equilibrium_wolf_algo
        self.Temperature = np.float(2.0)
        self.wolf = wolf*1
        M_neighboors = -1*np.ones((Graph.number_of_nodes(),Graph.number_of_nodes()),dtype=np.intc)
        M_neighboors = np.array(M_neighboors,dtype=np.intc)
        l_n_neighboors = np.zeros((Graph.number_of_nodes()))
        for node in range(Graph.number_of_nodes()):
            L_neigh = [n for n in Graph.neighbors(node)]
            l_n_neighboors[node] = len(L_neigh)
            if len(L_neigh)>0:
                M_neighboors[node,:len(L_neigh)] = np.array(L_neigh,dtype=np.intc)
        self.M_neighboors = M_neighboors
        self.l_n_neighboors = np.array(l_n_neighboors,dtype=np.intc)
        srandom(<unsigned int>time(NULL))
      
    
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.initializedcheck(False)
    def MSA_No_Phylogenie(self, char[:,:,::1] M_Different_Mutation_MSA_view):
        cdef:
            double[:,:,::1] L_spin = np.zeros((M_Different_Mutation_MSA_view.shape[0],M_Different_Mutation_MSA_view.shape[1],2))
            int index_MSA, index_average, flip
        for index_average in range(M_Different_Mutation_MSA_view.shape[0]):
            for index_MSA in range(M_Different_Mutation_MSA_view.shape[1]):
                if self.wolf==1:
                    for flip in range(self.Flip_equilibrium_wolf_algo):
                        self.Wolf_algo(M_Different_Mutation_MSA_view[index_average,index_MSA])
                else:
                    self.MCMC_adjency_cython_2(self.Flip_equilibrium_standard_algo, M_Different_Mutation_MSA_view[index_average,index_MSA])                
        
    @cython.cdivision(True)
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.initializedcheck(False)
    def MSA_of_Different_Number_Mutation(self, char[:,:,:,::1] M_Different_Mutation_MSA_view,
                                         list L_Mutation_per_generation, int Start_from_equilibrium):
   
        cdef:
            int index_number_mutation, index_average, flip
            int[:] L_mutation = np.array( L_Mutation_per_generation,dtype = np.intc)
            char[:,::1] M_Spin = np.random.randint(2,size=(M_Different_Mutation_MSA_view.shape[2],M_Different_Mutation_MSA_view.shape[3]),dtype=np.int8)*2-1
        for index_number_mutation in range(M_Different_Mutation_MSA_view.shape[0]):
            for index_average in range(self.Number_for_average):
                if Start_from_equilibrium !=0:
                    if self.wolf==1:
                        for flip in range(self.Flip_equilibrium_wolf_algo):
                            self.Wolf_algo(M_Spin[index_average])
                    else:
                        self.MCMC_adjency_cython_2(self.Flip_equilibrium_standard_algo, M_Spin[index_average])          
                M_Different_Mutation_MSA_view[index_number_mutation,index_average,0] = M_Spin[index_average]
        for index_number_mutation in range(L_mutation.shape[0]):
            self.Numb_Mutation_Per_Generation = L_mutation[index_number_mutation]
            self.Creation_Number_Average_of_MSA(M_Different_Mutation_MSA_view[index_number_mutation])
    
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.initializedcheck(False)
    def mcmc_matrix(self, char[:,::1] m_spin, int n_mutation):
        cdef int index_sequence
        for index_sequence in range(m_spin.shape[0]):
            self.MCMC_adjency_cython_2(n_mutation, m_spin[index_sequence])
    
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.initializedcheck(False)
    def mcmc_matrix_wolf(self, char[:,::1] m_spin, int n_mutation):
        cdef int index_sequence,i
        for index_sequence in range(m_spin.shape[0]):
            for i in range(n_mutation):
                self.Wolf_algo(m_spin[index_sequence])
                            
    @cython.cdivision(True)
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.initializedcheck(False)
    cdef void Creation_Number_Average_of_MSA(self, char[:,:,::1] M_MSA ) nogil:
        """
        Flip_equilibrium correspond to the number of flip before phylogeni start (hint : 0 correspond to start without equilibrium)
        Be Careful : Change parameter in Class !!!!
        """ 
        cdef int test,generation
        for test in range(self.Number_for_average):
            for generation in range(1,self.Number_Generation+1):
                M_MSA[test,int(2**(generation-1)):int(2**generation),:] = M_MSA[test,0:int(2**(generation-1)),:]
                self.One_Generation_More(M_MSA[test],int(2**generation))

    @cython.cdivision(True)
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.initializedcheck(False)
    cdef void One_Generation_More(self, char[:,::1] M_Spin, int index_last_generation) nogil:
        cdef int index_sequence
        for index_sequence in range(index_last_generation):
            self.MCMC_adjency_cython_2(self.Numb_Mutation_Per_Generation, M_Spin[index_sequence])
 
    @cython.cdivision(True)
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.initializedcheck(False)
    cdef void MCMC_adjency_cython_2(self, int Number_of_Mutation, char[::1] L_Spin) nogil:  
        cdef:
            int selected_node, c_mutation = 0
            float Prob 
        while c_mutation<Number_of_Mutation:   
            selected_node = int(drand48()*self.Number_of_Node)
            Prob = exp(-2.0/(self.Temperature)*self.Product_Spin(selected_node, L_Spin))
            if Prob>1.0 or drand48()<=Prob:
                L_Spin[selected_node]= -1* L_Spin[selected_node]
                c_mutation += 1

    @cython.cdivision(True)
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.initializedcheck(False)
    cdef float Product_Spin(self, int node, char[::1] L_Spin) nogil:
        cdef:
            int i
            float result = 0
        for i in range(self.l_n_neighboors[node]):
            result += L_Spin[self.M_neighboors[node,i]]
        return result*L_Spin[node]  
    
    @cython.cdivision(True)
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.initializedcheck(False)
    cdef void Wolf_algo(self, char[::1] L_spin):
        cdef:
            int[::1] L_node_cluster = -1*np.ones((self.Number_of_Node),dtype = np.intc)
            int selected_node = np.random.randint(0,self.Number_of_Node,dtype=np.intc)
            int counter_new_node = 1, counter_old_node = 0, new_node_cluster, node_around_new_node, node_test, save_counter_new_node, save_counter_old_node
            Py_ssize_t index_new_node_cluster,index_node_cluster
        L_node_cluster[0] = selected_node
        cdef float Proba_acceptance_in_cluster_wolf = (1-exp(-2.0/self.Temperature))
        while (counter_new_node - counter_old_node)!=0 :
            save_counter_new_node = counter_new_node
            save_counter_old_node = counter_old_node
            counter_old_node = counter_new_node
            for index_new_node_cluster in range(save_counter_old_node,save_counter_new_node):
                new_node_cluster = L_node_cluster[index_new_node_cluster]
                for index_node_around_new_node in range(self.l_n_neighboors[new_node_cluster]):
                    node_around_new_node = self.M_neighboors[new_node_cluster,index_node_around_new_node]
                    if L_spin[new_node_cluster]==L_spin[node_around_new_node] and self.item_in_liste(L_node_cluster[:counter_new_node],node_around_new_node)==0:                    
                        if drand48()<Proba_acceptance_in_cluster_wolf:
                            L_node_cluster[counter_new_node] = node_around_new_node
                            counter_new_node+=1   
        for index_node_cluster in range(counter_new_node):
            L_spin[L_node_cluster[index_node_cluster]] = -1*L_spin[L_node_cluster[index_node_cluster]]
            
    @cython.cdivision(True)
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.initializedcheck(False)
    cdef int item_in_liste(self, int[::1] List, int item) nogil:
        cdef int item_liste, index_item_liste
        for index_item_liste in range(len(List)):
            item_liste = List[index_item_liste]
            if item_liste == item:
                return 1
            if item_liste == -1:
                break
        return 0
    
    @cython.boundscheck(False)
    @cython.wraparound(False)
    @cython.initializedcheck(False)
    cdef double sum_array(self, char[::1] List) nogil:
        cdef:
            double c = 0
            int i
        for i in range(List.shape[0]):
            c += List[i]
        return c
    
```

```python
import numpy as np
from scipy.optimize import linear_sum_assignment

def Inference_Partner(M_MSA,Size_Training,regularisation, number_of_pair_in_species, middle_index = None, fast = False):
    """
    Input:
    middle_index = index which split in two protein the MSA
    """
    Percentage_True_partner = np.zeros(number_of_pair_in_species)
    counter_average = 0
    n_avg = 20
    for i in range(n_avg):
        for index_test in range(M_MSA.shape[0]):
            Liste_permutation = np.random.permutation(M_MSA.shape[1])
            MSA_Training = M_MSA[index_test,Liste_permutation[:Size_Training]]
            if fast :
                index_max = min(Size_Training + 40*number_of_pair_in_species,np.size(M_MSA,axis=1))
                MSA_Testing = M_MSA[index_test,Liste_permutation[Size_Training:index_max]]
            else :
                MSA_Testing = M_MSA[index_test,Liste_permutation[Size_Training:]]
            Jij = Inference_Jij(MSA_Training,regularisation)
            Eij = Energy_Partner(MSA_Testing,Jij,middle_index)
            ind_species = number_of_pair_in_species
            while ind_species<MSA_Testing.shape[0]:
                Cost = Eij[ind_species-number_of_pair_in_species:ind_species,ind_species-number_of_pair_in_species:ind_species]
                Permutation_worker_row = rng.permutation(Cost.shape[0])
                Cost = Cost[Permutation_worker_row]
                row_ind, col_ind = linear_sum_assignment(Cost)
                Percentage_True_partner += Permutation_worker_row[row_ind] == col_ind # The true index of worker i is Permutation_worker[i]
                counter_average += 1
                ind_species+=number_of_pair_in_species
    return np.mean(Percentage_True_partner/counter_average)

def Energy_Partner(MSA_Testing, Jij, middle_index = None):
    """
    Input :
    MSA_Testing
    Jij = Coupling inferred on MSA_training
    middle_index = index which split in two protein the MSA
    Output 
    Eij = Energy protein in MSA_L1 with protein in MSA_L2
    """
    if middle_index is None:
        middle_index = int(MSA_Testing.shape[1]/2)
    MSA_L1,MSA_L2 = MSA_Testing[:,:middle_index], MSA_Testing[:,middle_index:]
    M_conca_r = np.zeros((MSA_Testing.shape[0], MSA_L1.shape[1] + MSA_L2.shape[1]))
    M_conca_l = np.zeros((MSA_Testing.shape[0], MSA_L1.shape[1] + MSA_L2.shape[1]))
    M_conca_r[:,:MSA_L1.shape[1]] = MSA_L1
    M_conca_l[:,MSA_L1.shape[1]:] = MSA_L2
    Eij = - np.matmul(M_conca_l,np.matmul(Jij,M_conca_r.T))
    return Eij     
           
def Inference_Jij(MSA_Training,regularisation):
    Cij_regularized = regularisation_MSA(MSA_Training,regularisation) 
    Jij = - np.linalg.inv(Cij_regularized)
    return Jij
                
def regularisation_MSA(MSA_input,regularisation):
    """
    MSA : database of list of realisation of spin  [[one realisation],[a other], ] (Conversion from bool to -1 +1 if type is bool)
    regularisation : pourcentage of random chain of spin add to make Corelation matrix invertible (lambda)
    return : Correlation_Matrix regularized by if i!=j : C_ij = (1-lambda)<\sigma_i \sigma_j> - (1-lambda)**2<\sigma_i> <\sigma_j> = (1-lambda)*C_ij(sans regu) + ((1-lambda) - (1-lambda)**2) * Matrix(<\sigma_i><\sigma_j>)
    if i=j : C_ii = lambda + (1-lambda)<\sigma_i \sigma_i> - (1-lambda)**2<\sigma_i>**2 
    """
    if MSA_input.dtype == "bool":
        MSA = (MSA_input*1.0 - 0.5)*2.0
    else:
        MSA = MSA_input
    MSA = MSA.T
    avg_a = np.average(MSA,axis = 1)
    Vector_matrix = np.tile(avg_a,(len(avg_a),1))
    Id_avg = np.diagflat(avg_a)
    Matrix_product_avg = Id_avg@Vector_matrix
    Cij = np.cov(MSA)
    C_regularized = (1-regularisation)*(Cij+ regularisation* Matrix_product_avg) + regularisation*np.identity(len(avg_a))
    return C_regularized

def triangularise(m):
    """
    Put -inf in the diagonale and below it
    """
    m = np.asanyarray(m)
    mask = np.tri(*m.shape[-2:], k=0, dtype=bool)
    return np.where(mask, -1*np.ones(1, m.dtype)*np.inf, m)
```

### MI

```python
def Inference_Partner_Mutual_Info(M_MSA,Size_Training, pseudocount, number_of_pair_in_species, middle_index, fast = False):
    Percentage_True_partner = np.zeros(number_of_pair_in_species)
    counter_average = 0
    for index_test in range(M_MSA.shape[0]):
        Liste_permutation = np.random.permutation(M_MSA.shape[1])
        MSA_Training = M_MSA[index_test,Liste_permutation[:Size_Training]]
        if fast :
            index_max = min(Size_Training + 40*number_of_pair_in_species,np.size(M_MSA,axis=1))
            MSA_Testing = M_MSA[index_test,Liste_permutation[Size_Training:index_max]]
        else :
            MSA_Testing = M_MSA[index_test,Liste_permutation[Size_Training:]]
        if middle_index is None:
            middle_index = int(MSA_Testing.shape[1]/2)
        PMI_matrix = PMI_cython(MSA_Training, pseudocount, middle_index)
        ind_species = number_of_pair_in_species
        Cost = np.zeros((number_of_pair_in_species,number_of_pair_in_species))
        while ind_species<MSA_Testing.shape[0]:
            Cost = -1*Score_Sij_cython(Cost, MSA_Testing[ind_species-number_of_pair_in_species:ind_species], PMI_matrix, middle_index)
            Permutation_worker_row = rng.permutation(Cost.shape[0])
            Cost = Cost[Permutation_worker_row]
            row_ind, col_ind = linear_sum_assignment(Cost)
            Percentage_True_partner += Permutation_worker_row[row_ind] == col_ind # The true index of worker i is Permutation_worker[i]
            counter_average += 1
            ind_species+=number_of_pair_in_species
    return np.mean(Percentage_True_partner/counter_average)
```

```python
%%cython 
import numpy as np
cimport cython
from libc.math cimport log

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.initializedcheck(False)
def Score_Sij_cython(double[:,:] S_AB, char[:,:] MSA_Testing, double[:,:,:,:] PMI, int middle_index):
    cdef:
        int ind_liste,ind_liste_2,i,j,spin_i,spin_j
    for ind_liste in range(S_AB.shape[0]):
        for ind_liste_2 in range(S_AB.shape[1]):
            S_AB[ind_liste,ind_liste_2] = 0
            for i in range(middle_index):
                for j in range(middle_index,MSA_Testing.shape[1]):
                    spin_i = int((MSA_Testing[ind_liste,i]+1)/2)
                    spin_j = int((MSA_Testing[ind_liste_2,j]+1)/2)
                    S_AB[ind_liste,ind_liste_2] += PMI[i,j,spin_i,spin_j]
    return np.asarray(S_AB)

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.initializedcheck(False)
cpdef void two_body_freq_cython(double[:,:,:,::1] Fij, char[:,::1] MSA, double pseudocount, int middle_index):
    cdef:
        int site_i,site_j,index_try,index_spin_1,index_spin_2,spin_i,spin_j
        double delta_freq =  1.0/(MSA.shape[0])
    for site_i in range(middle_index):
        for site_j in range(middle_index,MSA.shape[1]):
            for index_try in range(MSA.shape[0]):
                spin_i = int((MSA[index_try,site_i]+1)/2.0)
                spin_j = int((MSA[index_try,site_j]+1)/2.0)  
                Fij[site_i,site_j,spin_i, spin_j ] += delta_freq
            for index_spin_1 in range(2):
                for index_spin_2 in range(2):
                    Fij[site_i,site_j,index_spin_1,index_spin_2] = pseudocount/4 + (1-pseudocount)*Fij[site_i,site_j,index_spin_1,index_spin_2]
            

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.initializedcheck(False)
def PMI_cython(char[:,::1] MSA,double pseudocount,int middle_index):
    cdef :
        int site_i,site_j,index_spin_1,index_spin_2,index_try
        float compteur = 0
        ##PMI[site_i,site_j,spin_i,spin_j]
        double[:,:,:,::1] PMI = np.zeros((MSA.shape[1],MSA.shape[1],2,2))
        double[:,:,:,::1] Fij = np.zeros((MSA.shape[1],MSA.shape[1],2,2))
        double[:,::1] Frequence_1body = np.zeros((MSA.shape[1],2))
    ##Cython
    two_body_freq_cython(Fij, MSA, pseudocount, middle_index)
    ##
    for site_i in range(MSA.shape[1]):
        compteur = 0
        for index_try in range(MSA.shape[0]):
            if MSA[index_try,site_i] == 0:
                compteur+=1
        Frequence_1body[site_i,0] = pseudocount/2 + (1-pseudocount)*compteur/MSA.shape[0]
        Frequence_1body[site_i,1] = 1 - Frequence_1body[site_i,0]
        
    for site_i in range(middle_index):
        for site_j in range(middle_index,MSA.shape[1]):
            for index_spin_1 in range(2):
                for index_spin_2 in range(2):
                    PMI[site_i,site_j,index_spin_1,index_spin_2] = log(Fij[site_i,site_j,index_spin_1,index_spin_2]/(Frequence_1body[site_i,index_spin_1]*Frequence_1body[site_j,index_spin_2]))
            
    return np.asarray(PMI)
```
