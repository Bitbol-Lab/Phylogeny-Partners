```python
%%cython --f
# cython: language_level=3, boundscheck=False, wraparound=False, initializedcheck=False
# distutils: extra_compile_args=-fopenmp
# distutils: extra_link_args=-fopenmp


import numpy as np
cimport numpy as np
cimport cython
from libc.stdlib cimport RAND_MAX, rand
from posix.stdlib cimport random, srandom
from libc.math cimport exp
from libc.time cimport time
from cython.parallel import prange
from libc.time cimport time


cdef class Creation_MSA_Generation_arDCA:
    """
    np.float64_[:,::1] Field,
                 np.float64_[:,:,:,::1] Coupling,
                 np.float64_[::1] p0,
                 long[::1] idxperm
    """
    cdef double[:,::1] Field
    cdef double[:,:,:,::1] Coupling
    cdef double[::1] p0
    cdef long[::1] idxperm
    cdef public int Number_state_spin
    cdef public int Number_of_Node
    cdef public double Temperature
    
    def __init__(self, Field,
                 Coupling,
                 p0,
                 idxperm):
        """
        """
        self.Number_of_Node = np.intc(Field.shape[0])+1
        self.Number_state_spin = np.intc(Field.shape[1])
        self.p0 = p0
        self.idxperm = idxperm -1 ## Convert to Python Style !!!!
        self.Field = Field
        self.Coupling = Coupling
        self.Temperature = 1
        srandom(<unsigned int>time(NULL))
        
    def msa_no_phylo(self, long n_sequences):
        cdef:
            int i,site,j,a,q,N
            char[::1] sample_z
            int[::1] backorder
            char[:,::1] msa
            double[::1] totH,p
        srandom(<unsigned int>time(NULL))
        q = self.p0.shape[0]
        N = self.Field.shape[0]
        msa = np.zeros((n_sequences,N+1),dtype=np.int8)
        backorder = np.array(np.argsort(np.asarray(self.idxperm)),dtype=np.intc)
        totH = np.zeros((q),dtype=np.double)
        sample_z = np.zeros((N+1),dtype=np.int8)
        for i in range(n_sequences):
            sample_z[<int>0] = rand_choice(self.p0)
            for site in range(N):
                totH[:] = self.Field[site]
                for j in range(site+1):
                    for a in range(q):
                        totH[a] += self.Coupling[site,a,sample_z[j],j]
                p = softmax(totH)
                sample_z[site + 1] = rand_choice(p)
            for j in range(sample_z.shape[0]):
                msa[i,j] = sample_z[backorder[j]]
        return np.asarray(msa)
                
                
    def msa_phylo(self, int n_generations, int n_mutations_generation, int start_equi):
        """
        int n_generations, int n_mutations_generation, int start_equi
        """
        cdef:
            char[::1] sequence 
            char[:,::1] msa = np.zeros((int(2**n_generations),self.Number_of_Node)
                                ,dtype = np.int8)
            int generation,index_sequence
        if start_equi==0:
            sequence = np.random.randint(0,high=self.Number_state_spin
                                 ,size = (self.Number_of_Node)
                                 ,dtype = np.int8)
        else:
            sequence = self.msa_no_phylo(1)[0]
        msa[0] = sequence
        for generation in range(1,n_generations+1):
            msa[int(2**(generation-1)):int(2**generation),:] = msa[0:int(2**(generation-1)),:]
            for index_sequence in range(int(2**generation)):
                self.mcmc(n_mutations_generation, msa[index_sequence])
        return np.asarray(msa)
    
    ### Use of Fast tree ######
    def msa_tree_phylo(self, clade_root, start_equi):
        """
        clade_root, start_equi
        """
        if start_equi==0:
            first_sequence = np.random.randint(0,high=self.Number_state_spin
                                 ,size = (self.Number_of_Node)
                                 ,dtype = np.int8)
        else:
            first_sequence = self.msa_no_phylo(1)[0]
        msa = []
        return self._msa_tree_phylo_recur(clade_root, first_sequence, msa)
    
    def _msa_tree_phylo_recur(self, clade_root, previous_sequence, msa):
        b = clade_root.clades
        if len(b)>0:
            for clade in b:
                #Mutation on previous_sequences
                new_sequence = np.copy(previous_sequence)
                n_mutations = int(clade.branch_length*new_sequence.shape[0])
                self.mcmc(n_mutations, new_sequence)
                self._msa_tree_phylo_recur(clade,new_sequence,msa)
        else:
            n_mutations = int(clade_root.branch_length*previous_sequence.shape[0])
            self.mcmc(n_mutations, previous_sequence)
            msa.append(previous_sequence)
        return msa

    #### standard function ########
    cpdef void mcmc(self, int Number_of_Mutation, char[::1] sequence):  
        cdef:
            int new_state, old_state, selected_node, c_mutation = 0
            double prob,prob_old,prob_new
        while c_mutation<Number_of_Mutation:   
            selected_node = randint(0,self.Number_of_Node)
            new_state = randint(0,self.Number_state_spin-1)
            old_state = sequence[selected_node]
            if new_state >= sequence[selected_node]:
                new_state += 1            
            prob_old = self.prob_sequence_cython(sequence)
            sequence[selected_node] = new_state
            if prob_old == 0:
                prob = 1
            else:
                prob_new = self.prob_sequence_cython(sequence)
                prob = prob_new/prob_old
            if random()/<float>RAND_MAX<prob:
                sequence[selected_node]= new_state
                c_mutation += 1
            else:
                sequence[selected_node]= old_state

    cdef double prob_sequence_cython(self,char[::1] sequence):
        cdef:
            int site, site1, a
            int q = self.p0.shape[0], N = self.Field.shape[0]
            double[:,::1] tottotH = np.zeros((N,q))
            double z, prob = self.p0[sequence[self.idxperm[0]]] ##Check if it is the good way ? 
        tottotH[:] = self.Field
        for site in prange(N, nogil = True):#Loop on index-1 already
            #totH[:] = self.Field[site] #copy memoryview
            for site1 in range(site+1): 
                for a in range(q):
                    tottotH[site,a] += self.Coupling[site,a,int(sequence[self.idxperm[site1+1]]),site1]#site1+1 because in reality N+1, here first choose done by p0
            z = sum_exp(tottotH[site])
            prob *= exp(tottotH[site,sequence[self.idxperm[site+1]]])/z
        return prob

cdef double sum_exp(double[::1] x) nogil:
    cdef : 
        int i
        double sumx = 0.0 
    for i in range(x.shape[0]):
        sumx += exp(x[i])
    return sumx

cdef double[::1] softmax(double[::1] x):
    cdef : 
        int i
        double sumx = 0.0, max_x = max_list(x)
    for i in range(x.shape[0]):
        x[i] = exp(x[i] - max_x)
        sumx += x[i]
    for i in range(x.shape[0]):
        x[i]/=sumx
    return x

cdef double max_list(double[::1] x):
    cdef:
        double max_x = x[<int>0]
        int i
    for i in range(1,x.shape[0]):
        if x[i]>max_x:
            max_x = x[i]
    return max_x

cdef inline int randint(int lower, int upper):
    """lower included, upper excluded"""
    return random() % (upper - lower ) + lower

cpdef int rand_choice(double[::1] p):
    cdef:
        int i
        float randn
        double c = 0.0
    randn = random()/<float>RAND_MAX
    for i in range(p.shape[0]):
        c += p[i]
        if randn<=c:
            return i
```

```python
import cython_code.analyse_sequence as an
import ccmpred.raw
from Bio import SeqIO

def convert_msa_number(file,order_amino):
    msa=[]
    for record in SeqIO.parse(file, "fasta"):
        l=[]
        for amino in record.seq:
            index = order_amino.find(amino)
            l.append(index)
        msa.append(l)    
    return np.array(msa,dtype=np.int8)

msa_no_phylo_arDCA_jul_0_02 = np.array(np.load("data/msa_arDCA__lambdaJ_0.002_lambdaH_0.0001.npy"),dtype=np.int8) - 1 

pseudocount = 0
theta = 0.8
theta_ccmgen = 0.8 ## Reweight sequence if hamming distance<theta
n_states = 21

order_amino_arDCA = "ACDEFGHIKLMNPQRSTVWY-"
file = "fasta_file/Concat_nnn_withFirst.fasta"
msa_compar_arDCA = convert_msa_number(file,order_amino_arDCA)
msa_compar_arDCA_unique = np.unique(convert_msa_number(file,order_amino_arDCA),axis=0)

## Frequence from generated data
#Fij_ardca,_ = an.two_body_freq_cython(msa_no_phylo_arDCA,pseudocount,n_states,theta)
Fij_ardca,_ = an.two_body_freq_cython(msa_no_phylo,pseudocount,n_states,0)
Fij_ardca_jul,_ = an.two_body_freq_cython(msa_no_phylo_arDCA_jul_0_02,pseudocount,n_states,0)

## Frequence form original data
Fij_in_ardca,_ = an.two_body_freq_cython(msa_compar_arDCA,pseudocount,n_states,theta)
Fij_in_ardca_unique,_ = an.two_body_freq_cython(msa_compar_arDCA_unique,pseudocount,n_states,theta)
#Fij_in_ardca_theta_0,_ = an.two_body_freq_cython(msa_compar_arDCA,pseudocount,n_states,0)
```
